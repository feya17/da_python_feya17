```python
# Сколько вакансий, которые вам нравятся?# Сколько вакансий для аналатика данных?
from collections import Counter
DA_vact = [row for row in vact_low if ('дата аналитик' in row) == True or ('data analyst' in row) == True or ('аналитик данных' in row) == True]
cnt = Counter(row for row in (DA_vact))
#cnt.most_common(10)
sum(cnt.values())
```
