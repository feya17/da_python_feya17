```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly as pl
import plotly.express as px
import plotly.graph_objs as go
import os
```

```python
dataset_path = os.path.join('Spotify_Youtube.csv')
Spotify_Youtube = pd.read_csv(dataset_path)
```

```python
df = pd.DataFrame(Spotify_Youtube.Artist.value_counts().nlargest(7))
```

```python
df
```

```python
artist = pd.DataFrame(Spotify_Youtube, columns = ['Artist','Track','Views','Likes','Comments'])
```

```python
artist.head()
```

```python
artist = artist.fillna(0)
```

```python
artist['Views'] = artist['Views'].astype(np.int64)
artist['Likes'] = artist['Likes'].astype(np.int64)
artist['Comments'] = artist['Comments'].astype(np.int64)
```

```python
artist.info()
```

```python
#топ 5 исполнителей по просмотрам
top_artist = artist.nlargest(5, 'Views')
fig = go.Figure(data=[go.Bar(x=top_artist['Artist'], y=top_artist['Views'])])
fig.update_yaxes(range=[0, 2200000000])
fig.update_layout(title='TOP-5 исполнителей')
fig.show()
```

```python
#распределение комментов по исполнителю
fig = px.histogram(artist, x='Comments', y = 'Artist',  nbins=10,  color_discrete_sequence=['pink'])
fig.update_layout(xaxis_title="Комменты", yaxis_title="Исполнители" , title='Количество комментов  по каждому исполнителю')

fig.show()
```

```python
#Какие фильмы Тарантино попали в список
Gorillaz = artist[artist['Artist'] == 'Gorillaz']
fig = px.bar(Gorillaz, x='Artist', y='Likes', color='Track', 
             color_continuous_scale='Inferno', )
fig.update_yaxes(range=[0, 20000000])
fig.update_layout(xaxis_title="Исполнитель", yaxis_title="Лайки" , title='Количество лайков за трек')

fig.show()
```
